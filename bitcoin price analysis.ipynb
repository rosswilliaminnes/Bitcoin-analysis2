```python
from datetime import datetime
```

```python
dates = [datetime.strptime(d,'%d/%m/%Y') for d in data['Date']]
```

```python
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15, 6
```

```python
plt.plot(dates,data['Price'])
plt.plot(dates,pd.rolling_mean(data['Price'],3))
plt.plot(dates,pd.rolling_mean(data['Price'],5))
plt.plot(dates,pd.rolling_mean(data['Price'],7))
plt.show()
```

```python
# plotting a linear fit to some points, example
# sample data
x = np.arange(10)
y = 5*x + 10 

# fit with np.polyfit
m, b = np.polyfit(x, y, 1)

plt.plot(x, y, '.') # this generates the dots on the below graph
plt.plot(x, m*x + b, '-')
```
